# Lecture 2

*Apache Spark* is an open-source distributed general-purpose cluster computing framework with in-memory data processing engine.



### Architecture
1. SparkSQL - a component of the Spark, which supports data requesting by using SQL queries
2. Spark Streaming - processing of streaiming data in real-time
3. MLib (machine learning) - scalable machine learning library, which provides a common machine learning algorithms, such as classification, regression, clustering, collaborative, filtering, frequent pattern mining and etc.
4. GraphX (graph) - library for parallel processing of graphs
5. Apache Spark



**Resilient Distributed Datasets (RDD)** - are data structur distributed across main memories of cluster's machines for further parralel processing

*Resillient* - RDDS are immutable  
*Distributed* - distributed across cluster  
*Dataset* - because it holds data  



### Types of operations:
- Transformations - intended to transform RDD to the new RDD by applying operations, such as *map*, *filter*, *join*, and etc. (**Lazy mode**)
- Actions - produce some values obtained as a result of some calculations on the RDD, ex.: count, reduce

Calculation of transformation takes place only when the *action* is invoked, and the result has to be returned to the driver program
