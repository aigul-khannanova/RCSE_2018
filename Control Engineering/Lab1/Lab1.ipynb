At the present time for linear authomatic control systems analysis and synthesize frequency methods are used widely. they are based on system frequency responces observation.  
To get system frequency responses it is necessary to replace variable $s$ in transfer function $W(s)$ with $jw$ ($j$ is imaginary unit). In placement result function $W(jw)$ is formed.

```python
x = np.linspace(0.1, 0.50,500)
```

```python
# Original function

def W(s):
    return (15*s + 3)/((10*s + 3)*s)

#Canculate W(x)
y = np.apply_along_axis(W,0,x)

#Draw it
plt.xlabel('w')
plt.ylabel('W(w)')
plt.title('W(w) - original function')
plt.grid(True)
plt.plot(x,y)
```

```python
W(-5)
```

```python
# Determine U(w) and V(w)

def U_w(w):
    return 15 / (100*np.power(w,2) + 9)

def V_w(w):
    return -(150*np.power(w,2) + 9)/(100*np.power(w,3) + 9*w)
```

```python
# Create plots for U_w and V_w

y1 = list(map(U_w,x))
y2 = list(map(V_w,x))


plt.subplot(2, 1, 1)
plt.xlabel('w')
plt.ylabel('U(w)')
plt.title('U(w)')
plt.grid(True)
#plt.yscale('log')
plt.plot(x, y1, )


plt.subplot(2, 1, 2)
plt.xlabel('w')
plt.ylabel('V(w)')
plt.title('V(w)')
plt.grid(True)
#plt.yscale('log')
plt.plot(x, y2, )
```

```python
#Рисую график зависимости мнимой части от вещенственной

plt.xlabel("U(w)")
plt.ylabel("V(w)")
plt.title("V(U)")
plt.grid()
plt.plot(y1, y2)
```

```python
# Determine A(w)

def A_w(w):
    return np.sqrt(np.power(U_w(w),2) + np.power(V_w(w),2))
```

```python
# Determine phi_w

def phi_w(w):
    return np.arctan(V_w(w) / U_w(w))
```

```python
# Determine L(w)

def L_w(w):
    return 20 * np.log10(A_w(w))
```

```python
# Draw A_w

y = list(map(A_w,x))

plt.xlabel('w')
plt.ylabel('A(w)')
plt.title('A(w) - amplitude-frequency function')
plt.grid(True)
#plt.yscale('log')
plt.plot(x,y, color = "blue",)
```

```python
y1 = list(map(phi_w,x))

# Draw phi_w
plt.subplot(2,1,1)
plt.xlabel('w')
plt.ylabel('phi(w)')
plt.title('phi(w) - imaginary-frequency function')
plt.grid(True)
plt.plot(x,y1, color = "red",)
```

```python
# Draw phi_w in log
plt.subplot(2,1,2)
plt.xlabel('w')
plt.ylabel('phi(w)')
plt.title('phi(w) - log imaginary-frequency function')
plt.grid(True)
#plt.yscale('log')
plt.plot(x,y1, color = "blue",)
```

```python
y2 = list(map(L_w,x))

# Draw L_w
plt.xlabel('w')
plt.ylabel('L(w)')
plt.title('L(w) - logarithmic function')
plt.grid(True)
#plt.yscale('log')
plt.plot(x,y2, color = "brown",)
```
