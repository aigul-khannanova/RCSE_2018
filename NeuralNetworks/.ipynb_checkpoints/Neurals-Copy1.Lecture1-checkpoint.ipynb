### Types of learning
Knowledge is a familiarity, awareness, or undestanding of someone or something, such as facts, information, descriptions, or skills, which is acquired through experience or education by perceiving, discovering, or learning.

Knowledge can be acquired from:
* *From expert (expert systems).*  - Applicable if well defined rules accessible with precise input data.
* *From data:*
  * **Supervised learning** - the system is supplied with a set of training examples consisting of inputs and corresponding outputs.
  > We know both input and output, but don't know their interconnections.
    
    - Supplies the neural network with inputs and the correct outputs (results).
    - We can estimate a error vector for certain input.
    - Response of the network to the inputs is measured. The weights are modified to reduce the difference between the actual and desired outputs.
  * **Unsupervised learning** - the system is supplied with tranining examples consisting only of inputs. It's required to discover appropriate outputs.
  > We know inputs and algorithms on it, but don't know its meaning.
  
    - The training set only consists of input patterns.
    - The neural network adjusts its own weights so that similar inputs cause similar outputs. The network identifies the patterns and differences in the inputs without any external assistance.

## Activation Function
Activation functions purposes are to convert a input signal of a node in a ANN to an output signal. If we dont apply a Activation function then the output signal would be a simple **linear function**. A linear equation is easy to solve but they are limited in ther complexity and have less power to learn complex functional mappings from data.

Neural-Networks are considered Universal Function Approximators. It means that they can compute and learn any function at all. Almost any process we can think of can be represented as a functional computation in Neural Networks.


Most popular activation functions:
#### Sigmoid activation function  
  - Vanishing gradient proble.
  - Output is not zero centered. It makes the gradient update go too far in different directions.
  - Sigmoids saturate and kill gradients.
  - sigmoids have slow convergence.
  
  
  
#### Hyperbolic Tangent function - Tanh
  - Output is zero centered because its range in (-1,1).
  - Optimization is easier in this method hence it is always preferred over Sigmoid function.
  - Still suffers from Vanishing gradient problem.
  <img src="https://cdn-images-1.medium.com/max/1600/0*VHhGS4NwibecRjIa.png"/>


#### ReLu - Rectified Linear units
Become very popular in the past couple of years. It was recently proved that it had *6 times improvement in convergence* from Tanh function.
  - Avoid vanishing gradient problem (Most of Deep Learning Models use ReLu nowadays).
  - ** Should be used only within Hidden layers of a Neural Network Model**.

## How Neural Network calculate and process

```python
import numpy as np
from math import e


inputs = np.array((0.5, 0.5, 0.2), dtype=float)
weights = np.array(((0.2, -0.3, 0.7),(0.5,0.4,-0.3)))
# Thresholds
T1, T2 = 0.5, -0.5

#Stigmoid activation function
y = lambda x: 1 / (1+pow(e,-x))

# Accumulate nodes
nodes_matrix = inputs * weights

# Calculate nodes values
# For first node gets its w(1,2,3)
S1 = nodes_matrix[:1].sum() - T1
S2 = nodes_matrix[:2].sum() - T2
                   
# Calculate neural outputs
y1 = y(S1)
y2 = y(S2)
                   
print("Neuron #1 value is:", y1)
print("Neuron #2 value is:", y2)
```

### Fundamentals of learning and training samples

- **Weights** - the strength or amplitude of a connection between two nodes, corresponding in biology pf the ampunt of influence the firing of one neuron has on another.
- **Training set** - a set of training patterns (p,s), where p - input; s - desired output; which we use to train our neural net.
- **Training** - the act of presenting the network with some samle data and modifying tghe weights to better approximate the desired function.

**Training pattern** is an input vector *p* with the components *x1*, *x2*, ..., *xn* whose desired output is known.  
By entering the training pattern into the network we receive an output that can be compared with the desired output.  
The set of training patterns is called **P**. It contains a finite number of ordered pairs *(p,r)* of training patterns with corresponding desired output *t*.

**The teaching input** *tj* is the desired and correct value *j* should output after after the input of certain trainging pattern.

**Error vector** - the difference between output vector and teaching input under a trining input *p*.


**Epoch** - single presentation of the entire data to the neural network. Typically many epochs are required to train the neural network.

**Iteration** - the process of providing the network with an single input and updating the network's weights
